### Reading an image

````python
import cv2
import matplotlib.pyplot as plt

image_path = './images/metro.jpg'
image = cv2.imread(image_path)
````

### Displaying and image
The openCV syntax is
```python
cv2.imshow('window name', image)
```
but the problem is that the image box is using the same Python process as the kernel. When you execute this from an interactive shell, the shell waits for you to get back from the image display box. But Jupyter wants to claim control back after you executed a command block. Therefore it is not happy to wait for the window, and you lose control of the Python process, crashing everything.
However, you still can only use it on your local machine and not in a server.

Matplotlib can also come in handy for displaying images. This is the solution used for displaying images on server.

````python
cv2.imshow('d',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(image)
````

#### Saving an image


````python
cv2.imwrite('./images/metro2.png',image)
plt.imshow(cv2.imread('./images/metro2.png'))
````

#### Accessing image properties
##### shape
When accessing the shape of the image, 
the first argument of the tuple is the height 
and the second is the width and the third one
is the number of channels that image has.

````python
print(image.shape)
````

#### Changing color-space
grayscale images are so important in image processing since they only have one channel.
this makes processing them easier in areas like contour detection.

````python
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
cv2.imshow('d',gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(gray)
````

As you see above, matplotlib shows the grayscale image with a yellowish kind of color
although opencv displays it correctly. to fix this issue, add this flag as the second argument to 
matplotlib. cmap="gray"

#### Color map
color map is a function that maps an image to another. To put it simple, it tells matplotlib how to process the image's matrix of pixels

````python
plt.imshow(gray, cmap="gray")
````

#### The other way around
It's impossible to convert a gray image to RGB.
You get a three channel again but the other channels are still empty because
you already lost other channels how can you retrieve that?


````python
new_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
plt.imshow(new_rgb)
````

#### Resizing image
By resizing it not only changes the dimensions, but also changes the resolution so as you see
we get such result. Quality of the image is lost.

````python
plt.imshow(cv2.resize(new_rgb, (100,100)))
````

#### Displaying text on images

````python
text = 'open cv'
font_scale = 4
thickness = 2
#first argument being width and second being height
text_pos = (100,500)
text_color = (0,255,255)
texted =  cv2.putText(new_rgb,text, text_pos ,
                      cv2.FONT_HERSHEY_SIMPLEX,
                      font_scale, text_color, thickness)
plt.imshow(texted)
````

#### Drawing
##### Line

````python
start_point = (0,0)
end_point = (1000,100)

plt.imshow(cv2.line(texted, start_point, end_point, (255,0,0), thickness))
````

##### Circle

````python
center_point = (500,500)
radius = 300
plt.imshow(cv2.circle(new_rgb, center_point, radius, (0,255,0) , thickness))
````

to prevent matplotlib from redrawing on the previous plot, we use this command
```python
plt.clf()
```

````python
plt.clf()
plt.imshow(new_rgb)
````

##### Rectangle

````python
start_rec_p = (50,50)
end_rec_p = (200,200)
plt.imshow(cv2.rectangle(new_rgb, start_rec_p, end_rec_p, (255,0,0), 50))
````

##### Ellipse

````python
center_coordinates = (120, 300)
axesLength = (100, 300)
angle = 100
# the whole idea about start and end angle is wether we want a complete ellipse or not. 
startAngle = 0
endAngle = 250
color = (255,255, 255)
thickness = 1
plt.imshow(cv2.ellipse(new_rgb, center_coordinates, axesLength,
           angle, startAngle, endAngle, color, thickness))
````
